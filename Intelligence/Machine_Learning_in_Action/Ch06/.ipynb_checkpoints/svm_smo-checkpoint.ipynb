```python
import svmMLiA
import numpy as np
```

```python
dataArr,labelArr = svmMLiA.loadDataSet('testSet.txt')
labelArr
```

```python
b,alphas = svmMLiA.smoSimple(dataArr,labelArr,0.6,0.001,40)
```

```python
print b
print alphas[alphas>0]
```

```python
# 支持向量的数量
print np.shape(alphas[alphas>0])

# 哪些数据点是支持向量
for i in range(100):
    if alphas[i]>0.0:
        print dataArr[i],labelArr[i]
```

```python
b,alphas = svmMLiA.smoP(dataArr,labelArr,0.6,0.001,40)
```

```python
# 利用得到的alphas矩阵分类
ws = svmMLiA.calcWs(alphas,dataArr,labelArr)
print ws
```

```python
datMat = np.mat(dataArr)
print datMat[0]*np.mat(ws)+b # 小于0，属于-1类；大于0，属于1类
```

```python
# 验证分类结果的正确性
print labelArr[0]
```

```python
# k是高斯径向基函数的衰减因子
# k小 ===》 减少每个支持向量的影响程度，因此需要更多的支持向量 
# 支持向量太多，相当于每次都利用整个数据集进行分类，就退化成kNN了
# 支持向量太少，则决策边界太差
reload(svmMLiA)
svmMLiA.testRbf(k1=1.1)
```
